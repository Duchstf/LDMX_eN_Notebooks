```python
gst_df = pd.read_hdf("/Users/wketchum/Data/LDMX/base_generation_test.hdf",key="gst_df")
```

```python
bins = np.arange(0,1.0,0.04)
var="ptl"
units="GeV/c"

query='index' #all
wght="MFP_pi"
sigma=1

#title="%s, $\\nu_\mu$ CC"%(fd_title)
xlabel="Final State Lepton $p_{T}$ (%s)"%(units)
ylabel="Events / %2.2f %s "%(bins[-1]-bins[-2],units)

df=gst_df.query(query)

fig,ax = plt.subplots()

plt.grid(axis='y', linewidth=0.5)
plt.hist(df[var], weights=df["wght"], color='red', label="CV", bins=bins,**pltops_hist)
plt.hist(df[var], weights=df["wght_%s_p%d"%(wght,sigma)], 
         color='blue', 
         label="%s +%d$\sigma$"%(wght,sigma), bins=bins,**pltops_hist)
plt.hist(df[var], weights=df["wght_%s_n%d"%(wght,sigma)], 
         color='green', 
         label="%s -%d$\sigma$"%(wght,sigma), bins=bins,**pltops_hist)
plt.legend()

#plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)

plt.show()
#plt.savefig("/Users/wketchum/Downloads/numucc_e_%s.png"%fd_leg)
```
