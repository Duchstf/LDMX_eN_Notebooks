## 2. Processor

```python
hists = {}

hists["lepton_pt"] = hist.Hist("Events", hist.Cat("sample", "Sample name"), 
                               hist.Bin("pt", r"Final State Lepton $p_{T}$ [GeV/c]", 20, 0, 1))
hists["lepton_e"] = hist.Hist("Events", hist.Cat("sample", "Sample name"), 
                               hist.Bin("e", r"Final State Lepton E [GeV]", 20, 0, 4))

hists["p1_pt"] = hist.Hist("Events", hist.Cat("sample", "Sample name"), 
                               hist.Bin("pt", r"Final State Proton $p_{T}$ [GeV/c]", 20, 0, 1.6))
hists["p1_e"] = hist.Hist("Events", hist.Cat("sample", "Sample name"), 
                               hist.Bin("e", r"Final State Proton E [GeV]", 20, 0, 4))
hists["p1_ke"] = hist.Hist("Events", hist.Cat("sample", "Sample name"), 
                               hist.Bin("ke", r"Final State Proton KE [GeV]", 20, 0, 4))

hists["n1_pt"] = hist.Hist("Events", hist.Cat("sample", "Sample name"), 
                               hist.Bin("pt", r"Final State Neutron $p_{T}$ [GeV/c]", 20, 0, 1.6))
hists["n1_e"] = hist.Hist("Events", hist.Cat("sample", "Sample name"), 
                               hist.Bin("e", r"Final State Neutron E [GeV]", 20, 0, 4))
hists["n1_ke"] = hist.Hist("Events", hist.Cat("sample", "Sample name"), 
                               hist.Bin("ke", r"Final State Neutron KE [GeV]", 20, 0, 4))
```

```python
def processor(file_name):
    
    gst_df = pd.read_hdf(file_name, key="gst_df")

    query='index' #all
    df=gst_df.query(query)
    
    #Varibale parameters
    wght="MFP_pi"
    sigma=1
    units="GeV/c"
    
    
    #--------------------LEADING LEPTONS
    #Define trigger
    l_trigger = df['ptl'] > 0.4
    
    #Lepton's pT > 0.4 GeV
    hists["lepton_pt"].fill(sample = 'CV', pt = df[l_trigger]['ptl'].to_numpy(), weight = df[l_trigger]["wght"].to_numpy())
    
    hists["lepton_pt"].fill(sample = "%s +%d$\sigma$"%(wght,sigma), pt = df[l_trigger]['ptl'].to_numpy(),
                            weight= df[l_trigger]["wght_%s_p%d"%(wght,sigma)].to_numpy())
    
    hists["lepton_pt"].fill(sample = "%s -%d$\sigma$"%(wght,sigma), pt = df[l_trigger]['ptl'].to_numpy(),
                            weight= df[l_trigger]["wght_%s_n%d"%(wght,sigma)].to_numpy())
    
    #Energy of lepton's pT > 0.4 GeV
    hists["lepton_e"].fill(sample = 'CV', e = df[l_trigger]['El'].to_numpy(), weight = df[l_trigger]["wght"].to_numpy())
    
    hists["lepton_e"].fill(sample = "%s +%d$\sigma$"%(wght,sigma), e = df[l_trigger]['El'].to_numpy(),
                            weight= df[l_trigger]["wght_%s_p%d"%(wght,sigma)].to_numpy())
    
    hists["lepton_e"].fill(sample = "%s -%d$\sigma$"%(wght,sigma), e = df[l_trigger]['El'].to_numpy(),
                            weight= df[l_trigger]["wght_%s_n%d"%(wght,sigma)].to_numpy())
    
    #--------------------LEADING PROTONS
    df['pt_p1'] = np.hypot(df['px_p1'], df['py_p1'])
    
    #Proton pt
    hists["p1_pt"].fill(sample = 'CV', pt = df[l_trigger]['pt_p1'].to_numpy(), weight = df[l_trigger]["wght"].to_numpy())
    
    hists["p1_pt"].fill(sample = "%s +%d$\sigma$"%(wght,sigma), pt = df[l_trigger]['pt_p1'].to_numpy(),
                       weight= df[l_trigger]["wght_%s_p%d"%(wght,sigma)].to_numpy())
    
    hists["p1_pt"].fill(sample = "%s -%d$\sigma$"%(wght,sigma), pt = df[l_trigger]['pt_p1'].to_numpy(),
                        weight= df[l_trigger]["wght_%s_n%d"%(wght,sigma)].to_numpy())
    #Proton Energy
    hists["p1_e"].fill(sample = 'CV', e = df[l_trigger]['E_p1'].to_numpy(), weight = df[l_trigger]["wght"].to_numpy())
    
    hists["p1_e"].fill(sample = "%s +%d$\sigma$"%(wght,sigma), e = df[l_trigger]['E_p1'].to_numpy(),
                       weight= df[l_trigger]["wght_%s_p%d"%(wght,sigma)].to_numpy())
    
    hists["p1_e"].fill(sample = "%s -%d$\sigma$"%(wght,sigma), e = df[l_trigger]['E_p1'].to_numpy(),
                        weight= df[l_trigger]["wght_%s_n%d"%(wght,sigma)].to_numpy())
    
    #Proton Kinetic Energy
    hists["p1_ke"].fill(sample = 'CV', ke = df[l_trigger]['ke_p1'].to_numpy(), weight = df[l_trigger]["wght"].to_numpy())
    hists["p1_ke"].fill(sample = "%s +%d$\sigma$"%(wght,sigma), ke = df[l_trigger]['ke_p1'].to_numpy(),
                       weight= df[l_trigger]["wght_%s_p%d"%(wght,sigma)].to_numpy())
    hists["p1_ke"].fill(sample = "%s -%d$\sigma$"%(wght,sigma), ke = df[l_trigger]['ke_p1'].to_numpy(),
                        weight= df[l_trigger]["wght_%s_n%d"%(wght,sigma)].to_numpy())
    
    #--------------------LEADING NEUTRONs
    df['pt_n1'] = np.hypot(df['px_n1'], df['py_n1'])
    
    #Proton pt
    hists["n1_pt"].fill(sample = 'CV', pt = df[l_trigger]['pt_n1'].to_numpy(), weight = df[l_trigger]["wght"].to_numpy())
    
    hists["n1_pt"].fill(sample = "%s +%d$\sigma$"%(wght,sigma), pt = df[l_trigger]['pt_n1'].to_numpy(),
                       weight= df[l_trigger]["wght_%s_p%d"%(wght,sigma)].to_numpy())
    
    hists["n1_pt"].fill(sample = "%s -%d$\sigma$"%(wght,sigma), pt = df[l_trigger]['pt_n1'].to_numpy(),
                        weight= df[l_trigger]["wght_%s_n%d"%(wght,sigma)].to_numpy())
    #Proton Energy
    hists["n1_e"].fill(sample = 'CV', e = df[l_trigger]['E_n1'].to_numpy(), weight = df[l_trigger]["wght"].to_numpy())
    
    hists["n1_e"].fill(sample = "%s +%d$\sigma$"%(wght,sigma), e = df[l_trigger]['E_n1'].to_numpy(),
                       weight= df[l_trigger]["wght_%s_p%d"%(wght,sigma)].to_numpy())
    
    hists["n1_e"].fill(sample = "%s -%d$\sigma$"%(wght,sigma), e = df[l_trigger]['E_n1'].to_numpy(),
                        weight= df[l_trigger]["wght_%s_n%d"%(wght,sigma)].to_numpy())
    
    #Proton Kinetic Energy
    hists["n1_ke"].fill(sample = 'CV', ke = df[l_trigger]['ke_n1'].to_numpy(), weight = df[l_trigger]["wght"].to_numpy())
    hists["n1_ke"].fill(sample = "%s +%d$\sigma$"%(wght,sigma), ke = df[l_trigger]['ke_n1'].to_numpy(),
                       weight= df[l_trigger]["wght_%s_p%d"%(wght,sigma)].to_numpy())
    hists["n1_ke"].fill(sample = "%s -%d$\sigma$"%(wght,sigma), ke = df[l_trigger]['ke_n1'].to_numpy(),
                        weight= df[l_trigger]["wght_%s_n%d"%(wght,sigma)].to_numpy())
    
    %reset -f in
```

```python
chunk = 0

for filename in os.listdir("../data/"):
    if filename.startswith("base_generation_with_weights_02Mar21_"):
        f_path = os.path.join("../data/", filename)
        processor(f_path)
        
        chunk += 1
#         if chunk == 1:
#             break
```
